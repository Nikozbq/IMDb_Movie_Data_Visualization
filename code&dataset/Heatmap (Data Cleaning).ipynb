```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib as plt
```

```python
df=pd.read_csv("top_5000_movies.csv")
df
```

```python
df["genre"]=df.genre.apply(lambda x:x.replace("\n","").replace(" ","").split(","))
df["genre"]
```

```python
from collections import Counter

# list of genres
genres=df["genre"].to_list()

# count the frequency of each genre
genre_counts = Counter(genre for movie in genres for genre in movie)

# initialize the correlation matrix
n = len(genre_counts)
correlation_matrix = [[0] * n for _ in range(n)]

# calculate the correlation between each pair of genres
for i, genre1 in enumerate(genre_counts):
    for j, genre2 in enumerate(genre_counts):
        if i < j:
            count = sum(1 for movie in genres if genre1 in movie and genre2 in movie)
            correlation = count
            correlation_matrix[i][j] = correlation
            correlation_matrix[j][i] = correlation

# print the correlation matrix
print('   ', end='')
for genre in genre_counts:
    print(f'{genre:<10}', end='')
print()
for i, row in enumerate(correlation_matrix):
    print(f'{genre_counts[i]:<2}', end='')
    for j, correlation in enumerate(row):
        print(f'{correlation:<10.2f}', end='')
    print()
```

```python
import csv

# list of genres and correlation matrix from previous example
genres = ["Action","Adventure","Fantasy","Comedy","Drama","Animation","Romance","War","Crime","Family","Sci-Fi","Biography","Thriller","Music","Horror","Mystery","History","Western","Sport","Musical","Film-Noir"]

# write the data to a CSV file
with open('genres.csv', 'w', newline='') as csvfile:
    writer = csv.writer(csvfile)
    writer.writerow([''] + genres)
    for i in range(len(genres)):
        row = [genres[i]] + [correlation_matrix[i][j] for j in range(len(genres))]
        writer.writerow(row)
```
