```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib as plt
```

```python
df=pd.read_csv("top_5000_movies.csv")
df["title"]
```

```python

df["title"]=df.title.apply(lambda x:x.replace("\n","").replace("0","").replace("1","").replace("2","").replace("3","").replace("4","").replace("5","").replace("6","").replace("7","").replace("8","").replace("9","").replace(".","").replace("(","").replace(")","").replace(",","").replace("&","").replace(":","").replace("-","").replace("Men","Man"))
titles=df["title"].to_list()
titles
```

```python
words=[]
for title in titles:
    tmp=title.split(" ")
    for i in range(0,len(tmp)):
        words.append(tmp[i])
print(len(words))
```

```python
df1=pd.read_csv("stop_words.csv")
stop_words=df1["Words"].to_list()
result=[]
for word in words:
    if word.lower() not in stop_words:
        result.append(word)
print(len(result))
```

```python
df2=pd.DataFrame(data=np.array(result),columns=["Word"])
word_count=df2.value_counts()
word_count
```

```python
index_list=[]
count=[]
for i in range(1,len(word_count)):
    index_list.append(word_count.index[i][0])
    count.append(word_count[i])
```

```python
import csv
with open('word_frequency.csv', 'w', newline='') as file:
    writer = csv.writer(file)
    writer.writerow(['Word', 'Count'])  # Write the header row
    for row in zip(index_list, count):
        writer.writerow(row)
```
