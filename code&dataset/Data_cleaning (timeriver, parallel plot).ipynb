```python
df["genre"]=df.genre.apply(lambda x:x.strip("[]").replace("'","").replace(" ","").split(","))
df["genre"]
```

```python
from collections import Counter

# list of genres
genres=df["genre"].to_list()

# count the frequency of each genre
genre_counts = Counter(genre for movie in genres for genre in movie)

# initialize the correlation matrix
n = len(genre_counts)
correlation_matrix = [[0] * n for _ in range(n)]

# calculate the correlation between each pair of genres
for i, genre1 in enumerate(genre_counts):
    for j, genre2 in enumerate(genre_counts):
        if i < j:
            count = sum(1 for movie in genres if genre1 in movie and genre2 in movie)
            correlation = count
            correlation_matrix[i][j] = correlation
            correlation_matrix[j][i] = correlation
genre=[]
# print the correlation matrix
print('   ', end='')
for g in genre_counts:
    print(f'{g:<10}', end='')
    genre.append(g)
print()
for i, row in enumerate(correlation_matrix):
    print(f'{genre_counts[i]:<2}', end='')
    for j, correlation in enumerate(row):
        print(f'{correlation:<10.2f}', end='')
    print()
```

```python
genres = genre
```

```python
raw=pd.read_csv("top_5000_movies.csv")
def int_year(data):
  if data[0] == '(':
    #print(data)
    return int(str(data).replace('(','').replace(')','').replace('I','').replace('X','')
    .replace(' ','').replace('V','').replace('TMovie','').replace('-','')[:4])
  else:
    return int(data)
def int_time(data):
  return int(data.strip(' min'))
raw['year'] = raw['year'].str.replace('-', '').apply(int_year)
raw['time'] = raw['time'].apply(int_time)
raw['rating'] = raw['rating'].apply(lambda x: float(x))
raw['metascore'] = raw['metascore'].apply(lambda x: int(x.strip(' \nMetascore')))
raw['countries'] = raw['countries'].replace(["Taiwan"],'China')
raw['countries'] = raw['countries'].replace(["Hong Kong"],'China')
raw['countries'] = raw['countries'].replace(["West Germany"],'Germany')
raw.head()
```

```python
df = raw
grouped = df.groupby('countries')['rating'].agg(['count', 'mean'])
grouped = grouped.rename(columns={'count': 'movie_count', 'mean': 'average_score'})
grouped = grouped.sort_values('average_score', ascending=False)
grouped.head()
```

```python
grouped.to_csv("cont_ra_cnt.csv",index = True)
```
